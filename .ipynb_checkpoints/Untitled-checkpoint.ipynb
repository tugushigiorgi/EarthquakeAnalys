```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from scipy.stats import norm
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Earthquakes_database.csv")

#lets see what we got here
df
```

```python
#lets split and have seperate columns with Year , Month and Day which willl help future for diagrams 

df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
   
df[['Year', 'Month', 'Day']] = df['Date'].apply(lambda x: pd.Series([x.year, x.month, x.day]))

 
df[['Year', 'Month', 'Day']] = df[['Year', 'Month', 'Day']].fillna(0).astype(int)

#lets drop Root mean squere column for example <we dont need 
df.drop('Root Mean Square', axis=1,inplace=True) 
#lets fill null values with not found 
df.fillna('Not Found', inplace=True)
df
```

```python
df["Magnitude"] 
```

```python

#lets group magnitude with Low , Medium and high and create new column with Magnitude Category 
df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=[0, 5, 7, 10], labels=['Low', 'Medium', 'High'])
 
```

```python

#this Csv file contains Time when accident happend , i think it will be better if i have seperate column
#which hold  Period of the day it happend like at Night , Day , or midnight based on Time 
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')


df = df.dropna(subset=['Time'])


time_bins = [0, 6, 18, 24]
time_labels = ['Midnight', 'Day', 'Night']

df['Period_of_Day'] = pd.cut(df['Time'].dt.hour, bins=time_bins, labels=time_labels, include_lowest=True, right=False)

 
df
```

```python

melted_df = pd.melt(df, id_vars=['Latitude', 'Longitude'], value_vars=['Type', 'Magnitude Type'])

#DataFrame after using melt
print("\nDataFrame after using melt:")
print(melted_df.head())

# Using pivot_table to summarize the data based on specific columns
pivot_table_df = pd.pivot_table(melted_df, values='value', index=['Latitude', 'Longitude'], columns='variable', aggfunc='first')

# DataFrame after using pivot_table
print("\nDataFrame after using pivot_table:")
print(pivot_table_df.head())
```

```python
 
#lets build some Cool charts :)
#Period of Day
#Depth
plt.figure(figsize=(12, 6))
plt.plot(df['Period_of_Day'], df['Depth'], label='Depth', marker='o', linestyle='-', color='blue')
plt.title('Depth Over Period')
plt.xlabel('Period of Day')
plt.ylabel('Depth')
plt.legend()
plt.show()

#Bar chart average magnitude by type for example earthquake , nuclear , rock burst etc..

plt.figure(figsize=(12, 6))
df_bar = df.groupby('Type')['Magnitude'].mean().reset_index()
plt.bar(df_bar['Type'], df_bar['Magnitude'], color='green')
plt.title('Bar Chart - Average Magnitude by Type')
plt.xlabel('Type')
plt.ylabel('Average Magnitude')
plt.show()

#scatter chart , magnitude seismic stations vs horizontal Distance 
    
plt.figure(figsize=(12, 8))
plt.scatter(df['Horizontal Distance'].dropna().astype(str), df['Magnitude Seismic Stations'].dropna().astype(str), c='red', alpha=0.5)
plt.title('Scatter Chart - Magnitude Seismic Stations vs Horizontal Distance')
plt.xlabel('Horizontal Distance')
plt.ylabel('Magnitude Seismic Stations')
plt.show()


#histogram , Depth Error distribution 


plt.figure(figsize=(12, 6))
plt.hist(df['Depth Error'].dropna().astype(str), bins=20, color='orange', edgecolor='black')
plt.title('Histogram - Depth Error Distribution')
plt.xlabel('Depth Error')
plt.ylabel('Frequency')
plt.show()



 
 
 
```
